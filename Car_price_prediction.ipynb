```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.tree import DecisionTreeRegressor
import statsmodels.api as sm
```

```python
df = pd.read_csv('df_cleaned.csv')
df.head()
```

```python
le = LabelEncoder()
def encoding(data, column):
    data[column] = le.fit_transform(data[column])
    return data
```

```python
cat_columns = df.select_dtypes('object')
for i in cat_columns:
    df = encoding(df,i)
```

```python
df.head()
```

```python
df.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
df.head()
```

## Model Build

###
- OLS

```python
# Creating the X and y variables
X = df.drop(['price'], axis=1)
y = df['price']
display(X.shape, y.shape)

# Adding a constant(intercept) to the independent variable
X_const = sm.add_constant(X)

# Fit the model
model = sm.OLS(y, X_const).fit()
display(model.summary())
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Assuming X includes all your independent variables (with constant)
# Create a DataFrame for VIF
vif_data = pd.DataFrame()
vif_data["feature"] = X_const.columns
vif_data["VIF"] = [variance_inflation_factor(X_const.values, i)
                   for i in range(X_const.shape[1])]

print(vif_data)
```

####
- Dropping some models with high variance inflation factor which shows a high multicolinearity, to help simplify the model

```python
# Dropping the features with high VIF and high p-values
X_const2 = X_const.drop(['highwaympg', 'curbweight', 'carlength', 'wheelbase', 'cylindernumber', 'aspiration'], axis=1)
y = df['price']

X_const2 = sm.add_constant(X_const2)

model2 = sm.OLS(y, X_const2).fit()
display(model2.summary())
```

```python
df['enginelocation'].value_counts()
```

```python
# dropping the enginelocation column as it has just one unique and therefore affects the simplicity of the model
# dropping other irrelevant features like enginetype (it is likely overshadowed by horsepower), carbody.
X_const3 = X_const2.drop(['enginelocation', 'enginetype', 'carbody', 'enginesize', 'fuelsystem'], axis=1)
y = df['price']
X_const3 = sm.add_constant(X_const3)

model3 = sm.OLS(y, X_const3).fit()
display(model3.summary())
```

###
- Linear Regression

```python
df.columns
```

```python
df.head()
```

```python
X = df.drop(['price', 'aspiration', 'carbody', 'enginetype', 'highwaympg', 'enginesize', 'fuelsystem', 'enginelocation', 'cylindernumber', 'curbweight', 'carlength', 'wheelbase'], axis=1)
y = df['price']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
display(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)
display(y_pred, y_test)
```

```python
# Checking the performance of the model
print('r2_score: ', r2_score(y_test, y_pred))
print('MAE: ', mean_absolute_error(y_test, y_pred)) 
print('MSE: ', mean_squared_error(y_test, y_pred))
```

```python
# Checking the intercept 
print('Intercept: ', lr.intercept_)
```

###
- Lasso Regression

```python
lr_2 = Lasso(alpha=0.1)
lr_2.fit(X_train, y_train)
y_pred_2 = lr_2.predict(X_test)
display(y_pred_2, y_test)
```

```python
# Checking the performance of the model
print('r2_score: ', r2_score(y_test, y_pred_2))
print('MAE: ', mean_absolute_error(y_test, y_pred_2))
print('MSE: ', mean_squared_error(y_test, y_pred_2))
# Checking the intercept
print('Intercept: ', lr_2.intercept_)
```

###
- Ridge Regression

```python
lr_3 = Ridge(alpha=0.1)
lr_3.fit(X_train, y_train)
y_pred_3 = lr_3.predict(X_test)
display(y_pred_3, y_test)
```

```python
# Checking the performance of the model
print('r2_score: ', r2_score(y_test, y_pred_3))
print('MAE: ', mean_absolute_error(y_test, y_pred_3))
print('MSE: ', mean_squared_error(y_test, y_pred_3))
# Checking the intercept
print('Intercept: ', lr_3.intercept_)
```

###
- DecisionTree Regressor

```python
lr_4 = DecisionTreeRegressor()
lr_4.fit(X_train, y_train)
y_pred_4 = lr_4.predict(X_test)
display(y_pred_4, y_test)
```

```python
# Checking the performance of the model
print('r2_score: ', r2_score(y_test, y_pred_4))
print('MAE: ', mean_absolute_error(y_test, y_pred_4))
print('MSE: ', mean_squared_error(y_test, y_pred_4))
```

###
- RandomForest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor()
rfr.fit(X_train, y_train)
y_pred_rfr = rfr.predict(X_test)
display(y_pred_rfr, y_test)
```

```python
# Checking the performance of the model
print('r2_score: ', r2_score(y_test, y_pred_rfr))
print('MAE: ', mean_absolute_error(y_test, y_pred_rfr))
print('MSE: ', mean_squared_error(y_test, y_pred_rfr))
```
